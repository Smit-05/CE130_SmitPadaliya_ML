*Stacking*

```python
import numpy as np
import pandas as pd

from sklearn.metrics import accuracy_score
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

#Models 
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import StackingClassifier
from sklearn.ensemble import AdaBoostClassifier
 
```

```python
cancer_data = load_breast_cancer()
X = cancer_data['data']
y = cancer_data['target']
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.5)
```

```python
model1 = LogisticRegression()
model2 = LinearRegression()
model3 = DecisionTreeClassifier()
```

```python
model1.fit(x_train,y_train)
model2.fit(x_train,y_train)
model3.fit(x_train,y_train)
```

```python
# Making dataset
pre1 = model1.predict(x_train)
pre2 = model2.predict(x_train)
pre3 = model3.predict(x_train)

test_pre1 = model1.predict(x_test)
test_pre2 = model2.predict(x_test)
test_pre3 = model3.predict(x_test)

track_stack = np.column_stack((pre1,pre2,pre3))
test_stack = np.column_stack((test_pre1,test_pre2,test_pre3))
```

```python
final_model = RandomForestClassifier()
final_model.fit(track_stack,y_train)
```

```python
final_pre = final_model.predict(test_stack)
```

```python
print(accuracy_score(y_test,final_pre))
```
