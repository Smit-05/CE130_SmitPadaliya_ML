```python
import numpy as np
import torch
```

```python
inputs = np.array([[73, 67, 43],
[91, 88, 64],
[87, 134, 58],
[102, 43, 37],
[69, 96, 70]], dtype='float32')
```

```python
targets = np.array([[56],
[81],
[119],
[22],
[103]], dtype='float32')
```

```python
inputs = torch.from_numpy(inputs)
targets = torch.from_numpy(targets)
print(inputs)
```

```python
weights = np.array([[0.5,1,1.5]],dtype='float32')
weights = torch.from_numpy(weights)
weights = torch.t(weights)
print(weights)
```

```python
multiplied = torch.mm(inputs,weights)
```

```python
bias = torch.tensor([[-2.2671,-2.2671,-2.2671,-2.2671,-2.2671]])
bias = torch.t(bias)
print(bias)
```

```python
prediction = torch.add(multiplied,bias)
print(prediction)
```

```python
sum = 0
for i in range(0,5):
  sum = sum + ((prediction[i]-targets[i])**2)
loss = sum/10
print(loss)
```
